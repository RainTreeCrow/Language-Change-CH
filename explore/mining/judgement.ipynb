Mannually examining the neighbourhood to judge if there has been a semantic change.

```python
from gensim.models.word2vec import Word2Vec

modeldir = '../../compass/5-year/model-cbow/'
models = []

for i in range(1945, 2025, 5):
    fiveyear = modeldir + str(i) + '-' + str(i + 4) + '.model'
    model = Word2Vec.load(fiveyear)
    models.append(model)
```

```python
def print_neighbourhood(word):
    year = 1945
    for model in models:
        string = str(year) + ': '
        for pair in model.wv.most_similar(word, topn=10):
            string += pair[0] + ', '
        print(string)
        year += 5
```
